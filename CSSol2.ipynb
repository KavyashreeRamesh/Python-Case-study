```python
#CASE STUDY2
import pandas as pd
import numpy as np
```

```python
store = pd.read_csv('Store.csv',sep=',',header=0,encoding = 'latin')
store
```

```python
#1 •	How many unique cities are the orders being delivered to
store.columns
#store['Segment'].unique()
```

```python
#2 •	What is the total quantity sold in the East Region?
e = store[store.Region == "East"]
e.groupby('Region').Quantity.count()
```

```python
#3 •	Find the sum of the quantity sold in the East Region
e = store[store.Region == "East"]
e.Quantity.sum()
```

```python
#4 •	In the south region sort the sales in decreasing order
s = store.loc[(store.Region == "South"), ['Region','Sales']]
s
s.sort_values(by=['Sales'],ascending = False)
```

```python
#5 •	Find the mean of quantity for every region
store.groupby('Region').Quantity.mean()
```

```python
#6 •	Find the mean of sales for every category
store.groupby('Category').Sales.mean()
```

```python
#7 •	Find the max, min, sum of sales and profit for every category
g = store.groupby('Category').Sales.agg([np.max,np.min,np.sum])
print("max, min, and sum of sales on each category :\n" ,g)

k = store.groupby('Category').Profit.max()
print("\nprofit of sales for each category:\n",k)
```

```python
#8 •	Find sum of sales and max profit for every segment
sum = store.groupby('Segment').Sales.sum()
print("max of sales for each seg:",sum)

max = store.groupby('Segment').Profit.max()
print("\n\nmax of profit for each seg :",max)
```

```python
#9 •	For every segment find the mean of the discount
store.groupby('Segment').Discount.mean()
```

```python
#10 •	For every segment find the most profitable customers
store.groupby('Segment').Profit.max()
store.loc[(store.Profit == 6719.9808) | (store.Profit == 8399.9760) | (store.Profit == 3919.9888) , ['Segment','Profit','Customer Name']]
```

```python
#11 •	What are the top 5 categories that give maximum profit?
l = store.sort_values(by=['Profit'],ascending=False)
l['Category'].head()
```

```python
#12 •	What is the Total Sales, Quantity, Discount, Profit across Total US.

store['Sales'].sum(),store['Quantity'].sum(),store['Discount'].sum(),store['Profit'].sum()
```

```python
#13 •	How many times has it taken more than 5 days from placing an order to shipping
from datetime import time
from datetime import date
from datetime import datetime
store['Order Date'] = pd.to_datetime(store['Order Date'])
store['Ship Date'] = pd.to_datetime(store['Ship Date'])
sdt = store['Ship Date'] - store['Order Date']
sdt
#store[sdt > '5 days'].shape[0]
```

```python
#14 •	Find the total number of orders in every category which has been shipped with a duration > 5 days
store['sdt'] = sdt
store[sdt > '5 days'].groupby('Category')['Order ID'].agg(['count'])
```

```python
#15•	What’s the percentage of items which has been shipped within 5 days
less = sdt < '5 days'
round((less.sum())/len(store)*100,3)
```

```python
#16•	What’s the percentage of items which has been shipped after 5 days
after = sdt > '5 days'
round((after.sum())/len(store)*100,3)
```
