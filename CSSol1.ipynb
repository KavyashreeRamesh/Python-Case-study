```python
#20 	Find:
#a. 	Mean and std deviation of the income
mean = oj.INCOME.mean()
std = oj.INCOME.std()
print("\nMean of income:" , mean , "\nStandard deviation of income :" , std)


#b. 	For income greater than or equal to 10.5, find the mean income
m = oj[oj.INCOME >= 10.5]['INCOME'].mean()
print("\nMean of Income for income >= 10.5:" ,m)


 #c. 	For each brand having price >=2.5 find the mean, median, sd of the log of income
a = oj[oj.price >= 2.5]
a
pmean =  a.groupby(oj['brand']).INCOME.mean()
print("\nmean of price for each brand :\n",pmean)

pmed = a.groupby(oj['brand']).INCOME.median()
print("\nmedian of price for each brand:\n",pmed)

pstd = a.groupby(oj['brand']).INCOME.std()
print("\nstandard deviation of price for each brand:",pstd)

plog = a.groupby(oj['brand']).log
```
